```python
# A list with strings,integers and boolean values:
list1 = ["abc",34,True,40,"male"]
print(list1)
```

```python
# Access Items
list1 = ["abc",34,True,40,"male"]
print(list1[0])
print(list1[2])
```

```python
# Negative/reverse indexing
thislist = ["apple","banana","cherry"]
print(thislist[-1])#here this is revesrse indexing and directly we reach the the last item in this last is -1 than -2 than -3 and so on
print(thislist[-3])
```

```python
#Slicing
thislist = ["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[2:5])# the syntax says 2 is inclusin-ve and visible in output and 5 is exclusive so uske ek pehele tak pas print hoga
print(thislist[:4])# is we
print(thislist[-3:])#empty means till the end and in revese we cant take 0 and if we take -1 it becomes exclusive so in order to proint last three we have to keep it empty
print(thislist[2:])
```

```python
#Change item value
thislist = ["apple","banana","cherry"]
thislist[1] = "blackcurrant"#replaceing of things in the original list
thislist[0]="watermelon"
print(thislist)
```

```python
# Append Items
thislist = ["apple","banana","cherry"]
thislist.append("orange")
print(thislist)
```

```python
#Insert Items
thislist = ["apple","banana","cherry"]
thislist.insert(2,"orange")# in append adds at last but here it is our choice
print(thislist)
```

```python
# Extend List
thislist.extend(list1)#note here two lists are added to make one list also list1 was from before so it takes it from memory
print(thislist)
```
